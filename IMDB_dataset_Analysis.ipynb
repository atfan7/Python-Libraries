# Question 5: In Gender column turn 'Male'=1 and 'Female'=0

```python
def encode (data):
  if data=='Male':
    return 1
  else:
    return 0

data['gender']= data['gender']. apply(encode)

data
```

# Question 6 : Create a new column called profit

```python
def prof(x):


  return x['revenue'] - x['budget']

data['profit'] = data[ ['revenue','budget']    ].apply(prof,axis=1)

data
```

# Question 12 : Find details of movies by high budget directors (HBD)

HBD>> atleast one movie with budget > 100M

```python
data['budget']= data['budget'] / 1000000
data['revenue'] = data['revenue'] / 1000000

data
```

```python
data_dir_budget = data.groupby('director_name')['budget'].max()

data_dir_budget
```

```python
data_dir_budget = data.groupby('director_name')['budget'].max().reset_index()
data_dir_budget
```

```python
names = data_dir_budget.loc [data_dir_budget['budget']>=100,'director_name']
names
```

```python
data.loc[data['director_name'].isin(names)]
```

Shortcut for Q11

```python
def func(x):

  return x['budget'].max ()>=100

data.groupby('director_name').filter(func)
```

or

```python
data.groupby('director_name').filter( lambda x : x['budget'].max()>=100 )
```

# Question 13 : Find all the risky movies.

Risky movie : if budget of new movie > avg revenue of old movies

```python
def funct(x):
  x['risky']= (x['budget'] - x['revenue'].mean()) >=0
  return x


data.groupby('director_name').apply(funct)
```
