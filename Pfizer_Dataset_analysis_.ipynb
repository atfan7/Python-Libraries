# Q1 : Melt the data frame i.e. Convert wide to long form of data

```python
pd.melt(data,id_vars=['Date','Drug_Name','Parameter'])
```

# Q2 : Change names of new created columns (variable and value)

```python
data_melt= pd.melt(data,id_vars=['Date','Drug_Name','Parameter'], var_name='Time', value_name='Readings')
data_melt
```

```python
data_melt.pivot
```

# Q4: Convert the long data to more wider data

```python
data_tidy = data_melt.pivot(index=['Date','Drug_Name','Time'],columns ='Parameter' ,values ='Readings').reset_index()
data_tidy
```

# Q9: Use Groupby for each column and put the mean of values for 'Null'

```python
def temp_mean(x):
  x['temp_avg']=x['Temperature'].mean()
  return x



data_tidy= data_tidy.groupby('Drug_Name').apply(temp_mean)

data_tidy
```

```python
data_tidy['Temperature'].fillna(data_tidy['temp_avg'],inplace =True)
```

```python
data_tidy.isna().sum()
```

# Q10: Binnig the temperature into 4 catagories : low(5-20),medium(20-35),high(35-50),very_high(50-60)

Numerical to Catagorical *data*

```python
temp_points=[5,20,35,50,60]
temp_lables= ["low","medium","high","very_high"]
```

```python
data_tidy['Temperature'] = pd.cut(data_tidy['Temperature'],bins=temp_points,labels = temp_lables)
data_tidy
```
